##4. 자율주행분야 : 자율주행 자동차 
* 자율주행 자동차란 운전자 또는 승객의 조작 없이 자동차 스스로 운행이 가능한 차를 말함
* 자율주행의 개념은 1960년대 벤츠를 중심으로 제안되었고, 1970년대 중후반부터 초보적인 수준의 연구가 시작되었음. 차세대 자동차 산업으로 주목받고 있는 기술
* 수많은 자동차 회사들 뿐만 아니라 구글, 애플 등 IT 기업들이 기술개발에 앞장서고 있음
* 개발현황
 * 현재 자율주행차 중 가장 앞서있다고 평가되는 것은 구글과 테슬라
  * 국내에서 자율주행차가 최초로 선보인 것은 1993년 대전엑스포때이며, 해당 차량은 세계 최초로 공도 주행을 한 차임. 서울 청계천~63빌딩까지 17km의 거리를 조작없이 성공적으로 운행하였으나, 해당 연구진은 해산하고 기술 또한 실전된 상태
 * 테슬라는 2019년 말까지 가장 많이 자율주행차를 판매한 개발 회사로서 테슬라의 모든 차량은 네트워크로 연결되어 다른 기업에 비해 많은 주행 데이터를 확보하고 있다고 평가됨
* 2012년 발표된 IEEE의 보고서에 의하면 2040년에는 전 세계 차량의 약 75%가 자율주행 자동차로 전환될 것으로 예상
* 상용화되면 전체 교통사고의 95%가량을 차지하는 운전자 부주의, 보복운전 등에 의한 사고를 줄일 수 있을 것으로 기대. 인간의 사망원인 중 교통사고가 차지하는 비율이 26.1%에 달하여 자율주행으로 교통사고를 줄이리 수 있다면 인간의 평균 수명은 더 길어질 것으로 예상
* 하지만 자율주행차에 대한 논의도 끊이지 않는 상황
 * 운수업계의 택시, 화물차, 버스운전자 등 여러 운전직 일자리가 감소하거나 심지어 사라질 전망으로 기계와 인간의 공존 방법의 모색이 필요
 * 주로 거론되는 문제는 "정말로 안전한지"임. 누군가 해킹으로 의도적으로 동시다발적으로 사고를 일으킬 수도 있고,자율주행차가 무인주행 도중에 사고가 발생하였을 경우 사고의 주체가 운전자가 아닌 "차"가 되므로 사고 발생 시 운전자와 자동차 생산업체 간의 소송이 발생 소지 높음
* 여러 가지 다양하고 복잡한 상황 가정 속에서도 자율주행차는 조만간 상용화 될 것이 분명하며, 인간은 다양한 변수들에 대한 해결책을 대비하며 이를 받아들일 준비가 필요하다.
